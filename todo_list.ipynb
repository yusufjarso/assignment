```python
task = ["cooking Food", "washing Cloth"]

index = 0
Menu ={
    1: "Add Task",
    2: "Task Status",
    3: "List Task",
    4: "Remove Task",
    5: "Quit",
    } 

def show_menu(Menu):
    print('TO_Do LIST MANAGER')
    print("\x1B[3mPlease welcome to the Task manager\x1B[0m \t \n")
    print('Menu:')
    print('1. Add Task')
    print('2. Task Status: \n \t a. Completed \n \t b. Not completed')
    print('3. List Tasks')
    print('4. Remove Task')
    print('5. Quit \n')
    return str(Menu)

show_menu(Menu)


selection = int(input("Please select the Menu Choices from  Numbers 1 - 5"))
if selection == 1:
    new_task = input("please add new task here!")
    task.append(new_task)
    print(f"New task of {new_task} was added. \nAll the tasks to be done are: \n {task} ")
   
# Check if the task was completed? if yes, mark as completed
    
elif selection == 2:
    user_input = input('Did you complete the task? (Y/N): ').lower()
    if user_input.lower() == 'y':
        completed_task = input("Which task did you complete?").lower()
        if completed_task in task:
            print(f'\n The task of {completed_task} was completed!')
        else:
            print(f'{completed_task} not in the task list. Please enter the correct task from the list of the task that you completed!')
    elif user_input.lower() == 'n':
            print(f"The task of {task} not yet completed!")
    else:
            print('Wrong selection! .....please type yes or no')
            
            
# List all the task in the To-Do list manager
            
elif selection == 3:
            print(f"The task to be completetd are: \n {task}")

elif selection == 4:
    choice = input(f"are you sure you want to delete any of {task} from the task? (Y/N) : ").lower()
    
    if (choice=="y"):
        remove_task = input("Please enter the task you want to delete?").lower()
        if remove_task in task:
            task.remove(remove_task)
            print(f" {Menu[4]} : \n \t The task of {remove_task} has been succesffuly removed from the task list!")
            print(f"\n You have only {task} to do!")
        else:
             print(f'{Menu[4]}: \n \t Please enter the correct task from the list of the task to be deleted!')
    elif (choice=="n"):
        print("No task has been deleted!. Please complete all the task")
    else:
        print('Wrong selection!.........  please type yes or no')
elif selection ==5:
    quit()
else:
    print('Invalid option. Please enter a number between 1 and 5.')
    
```
